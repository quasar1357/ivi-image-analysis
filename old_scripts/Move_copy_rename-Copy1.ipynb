# Renaming files in folders and subfolders

```python
import os
import re
import shutil
```

# copy, rename, move simultaneously

```python
dir_name = "./031_bAEC_IFA"
destination = "./data"

print(os.path.realpath(dir_name))
print(os.path.realpath(destination))
```

```python
for root, dirs, files in os.walk(dir_name, topdown=False):
    for file in files:
        full_file_path_0 = os.path.join(root, file)
        print(full_file_path_0)
```

```python
for root, dirs, files in os.walk(dir_name, topdown=False):
    for file in files:
        full_file_path = os.path.join(root, file)
        #full_file_path_0 = os.path.join(root, file)
        #full_file_path = full_file_path_0.split("/") [-1]
        
        
        if full_file_path.endswith(".tif"):
            channel_0 = full_file_path.split("/") [-1]
            channel_1 = re.findall('.+_.+_.+_.+_(.+)_.+\.tif', channel_0)[0]
            channel = channel_1.replace(" ", "")
            
            image_1 = full_file_path.split("/")[-2]
            image = re.findall('_.+_(.+)', image_1)[0]
            
            infection_1 = full_file_path.split("/")[-2]
            infection = re.findall('_(.+)_.+', image_1)[0]
            
            virus_1 = full_file_path.split('/')[-3]
            virus = re.findall('.+_(.+)_.+', virus_1)[0]
            
            temperature_1 = full_file_path.split("/")[-3]
            temperature = re.findall('.+_.+_(.+)', temperature_1)[0]
            
            donor_1 = full_file_path.split("/")[-3]
            donor = re.findall('(.+)_.+_.+', donor_1)[0]
            
            new_name = ("/" + donor + "_" + temperature + "_" + virus + "_" + infection + "_" + image + "_" + channel + ".tif")
            print(new_name)
            
            shutil.copy(full_file_path, destination + new_name)
            
            
```
